- High damage (H): $>=50\%$ roads blocked, or $>=30\%$ fires; 
- Medium damage (M): $[10\%, 50\%)$ roads blocked, or $[10\%, 30\%)$ fires; and 
- Low damage (L): $<10\%$ roads blocked, or $<10\%$ fires. 

Either blocked roads or fires that reach the range can be considered its damage level. For example, a region with 30\% road blocks and 5\% fires is Medium damaged.

- R1 (H): 60\% roads blocked, 35\% fires;

- R3 (M): 15\% roads blocked, 5\% fires; 
- R4 (H): 35\% roads blocked, 30\% fires; and 
- R5 (L): 5\% roads blocked, 3\% fires.

Assume all regions are of equal size to simplify the modelling process. Each region is modelled as a **node** in a graph, and predefined **distances** between regions (e.g., 5 km between R1 and R2, 3 km between R2 and R3) are detailed: R1 -- R2: 5 km, R1 -- R3: 7 km, R1 -- R4: 4 km, R1 -- R5: 6 km, R2 -- R3: 3 km, R2 -- R4: 4 km, R2 -- R5: 8 km, R3 -- R4: 5 km, R3 -- R5: 6 km, and R4 -- R5: 4 km.

If a fire truck reduces fire percentage by 10\%, that 10\% decrease is applied to the overall fire damage in that region, e.g., $20\%-10\%=10\%$. 

**Assumption:** Both rescue operations and disaster events (fire spread and aftershocks) do not consume time in this simulation, but each unit can only perform rescue once when visiting the region. #### Ant Colony Optimization (ACO) for Multi-Agent Coordination
##### Purpose: ACO simulates the behavior of agents (fire trucks and police units) navigating regions and responding to road blockages and fire spread using pheromone trails to mark visited paths. ACO models how multiple rescue agents (e.g., fire trucks and police) coordinate across paths while considering updated conditions.
##### Impact: By using pheromone levels that decay over time (simulating worsening conditions), ACO helps decide where rescue agents should prioritize actions. For instance, fire trucks follow paths where fire spread is more severe, and police units focus on areas with high road blockages.

```python
import numpy as np
# 1. evaporate_pheromones
def evaporate_pheromones(pheromone_levels, decay_rate=0.1):
    """
    Evaporates pheromone levels over time to simulate fire spread or aftershocks.
    """
    return {edge: level * (1 - decay_rate) for edge, level in pheromone_levels.items()}

# 2. deposit_pheromones
def deposit_pheromones(pheromone_levels, edge, units, agent_type="fire"):
    """
    Deposits pheromones based on the type of rescue unit.
    'fire' units reduce fire damage, 'police' units reduce road blockage.
    """
    if agent_type == "fire":
        reduction = units * 0.1  # Each fire truck reduces fire by 10%
    elif agent_type == "police":
        reduction = units * 0.1  # Each police unit reduces blockage by 10%
    
    # Update pheromone level based on rescue action
    pheromone_levels[edge] = max(0, pheromone_levels.get(edge, 1) - reduction)
    return pheromone_levels

# 3. choose_next_region
def choose_next_region(current_region, pheromone_levels, distances, regions, alpha=1, beta=2, agent_type="fire"):
    """
    Selects the next region to move to based on pheromone levels and distances.
    """
    probabilities = {}
    neighbors = [edge[1] for edge in distances if edge[0] == current_region] + \
                [edge[0] for edge in distances if edge[1] == current_region]
    
    for neighbor in neighbors:
        edge = (current_region, neighbor) if (current_region, neighbor) in distances else (neighbor, current_region)
        if agent_type == "fire":
            fire_pheromone = 1 - (regions[current_region]["fires"] + regions[neighbor]["fires"]) / 200
            pheromone = pheromone_levels.get(edge, fire_pheromone)
        elif agent_type == "police":
            blockage_pheromone = 1 - (regions[current_region]["roads_blocked"] + regions[neighbor]["roads_blocked"]) / 200
            pheromone = pheromone_levels.get(edge, blockage_pheromone)
        
        eta = 1 / distances[edge] if distances[edge] > 0 else 0
        probabilities[neighbor] = (pheromone ** alpha) * (eta ** beta)
    
    if not probabilities:
        return current_region
    
    total = sum(probabilities.values())
    probabilities = {k: v / total for k, v in probabilities.items()}
    return np.random.choice(list(probabilities.keys()), p=list(probabilities.values()))

# 4. edge_travel_time
def edge_travel_time(region_a, region_b, regions, max_speed=60):
    avg_blockage = (regions[region_a]["roads_blocked"] + regions[region_b]["roads_blocked"]) / 2
    speed_reduction = 1 - (avg_blockage / 100)
    travel_speed = max_speed * speed_reduction
    return distances[(region_a, region_b)] / travel_speed

# 5. simulate_agent_path
def simulate_agent_path(start_region, pheromone_levels, distances, regions, units, agent_type="fire"):
    """
    Simulates an agent's movement through the regions, updating pheromones based on actions.
    """
    current_region = start_region
    path = [current_region]
    
    while len(path) < len(regions):
        next_region = choose_next_region(current_region, pheromone_levels, distances, regions, agent_type=agent_type)
        if next_region == current_region:
            break
        edge = (current_region, next_region)
        pheromone_levels = deposit_pheromones(pheromone_levels, edge, units, agent_type=agent_type)
        path.append(next_region)
        current_region = next_region
    
    return path, pheromone_levels

# Sample pheromone levels initialized to 1 for all edges
pheromone_levels = {edge: 1 for edge in distances.keys()}

# Test ACO simulation with fire trucks and police units starting at R1
fire_truck_path, updated_pheromones_fire = simulate_agent_path(
    start_region="R1",
    pheromone_levels=pheromone_levels,
    distances=distances,
    regions=regions,
    units=2,
    agent_type="fire"
)

police_path, updated_pheromones_police = simulate_agent_path(
    start_region="R1",
    pheromone_levels=pheromone_levels,
    distances=distances,
    regions=regions,
    units=3,
    agent_type="police"
)

print("Fire Truck Path:", fire_truck_path)
print("Police Path:", police_path)
```

## Genetic Algorithms for Path Optimization

#### Genetic Algorithm (GA) for Path Optimization
##### Purpose: The GA determines optimal paths for rescue agents to minimize travel time across regions. It seeks paths that balance the need to visit all regions within 90 minutes while accounting for blocked roads that slow down travel.
##### Impact: This task provides a strategic path that rescue agents should follow to ensure each region is visited once and within the time constraint. It optimizes the sequence of visits, helping minimize delays caused by detours around heavily blocked regions.

```python
# 1. generate_initial_population
def generate_initial_population(aco_paths, num_paths, regions):
    """
    Initializes a population for the genetic algorithm.
    Starts with ACO paths and adds random variations if additional paths are needed.
    """
    population = [path for path in aco_paths]  # Start with ACO-generated paths
    while len(population) < num_paths:
        path = ["R1"] + list(np.random.permutation(list(regions.keys())[1:]))
        population.append(path)
    return population

# 2. fitness_function
def fitness_function(path, distances, regions):
    """
    Calculates fitness of a path based on travel time, which is influenced by road blockages.
    """
    travel_time = 0
    for i in range(len(path) - 1):
        region_a, region_b = path[i], path[i + 1]
        avg_blockage = (regions[region_a]["roads_blocked"] + regions[region_b]["roads_blocked"]) / 2
        speed_reduction = 1 - (avg_blockage / 100)
        travel_speed = 60 * speed_reduction
        travel_time += distances.get((region_a, region_b), distances.get((region_b, region_a), 0)) / travel_speed
    return 1 / (travel_time + 1)  # Higher fitness for lower travel time

# 3. selection
def selection(population, fitness_scores, selection_size=4):
    """
    Selects the top paths based on fitness scores for the next generation.
    """
    selected_indices = np.argsort(fitness_scores)[-selection_size:]
    return [population[i] for i in selected_indices]

# 4. crossover
def crossover(parent1, parent2):
    """
    Performs crossover by combining portions of two paths to create a new path.
    Ensures no duplicate regions and maintains a valid path.
    """
    start, end = sorted(np.random.choice(range(1, len(parent1)), 2, replace=False))
    child = parent1[:start] + parent2[start:end] + parent1[end:]
    missing = list(set(parent1[1:]) - set(child[1:]))
    for i in range(1, len(child)):
        if child.count(child[i]) > 1:
            if missing:
                child[i] = missing.pop(0)
    return child

# 5. exchange
def exchange(path):
    """
    Exchanges two segments of a path to simulate mutation.
    """
    path = path[:]
    i, j = np.random.choice(range(1, len(path)), 2, replace=False)
    path[i], path[j] = path[j], path[i]
    return path

# 6. successor
def successor(path):
    """
    Generates a new path configuration by applying an exchange operation to path[1:].
    """
    return [path[0]] + exchange(path[1:])

# 7. genetic_algorithm
def genetic_algorithm(aco_paths, num_generations, num_paths, distances, regions):
    """
    Main Genetic Algorithm function that applies selection, crossover, and exchange iteratively.
    """
    population = generate_initial_population(aco_paths, num_paths, regions)
    for generation in range(num_generations):
        fitness_scores = [fitness_function(path, distances, regions) for path in population]
        selected = selection(population, fitness_scores)
        next_generation = selected[:]
        while len(next_generation) < num_paths:
            parent_indices = np.random.choice(len(selected), 2, replace=False)
            parent1, parent2 = selected[parent_indices[0]], selected[parent_indices[1]]
            child = crossover(parent1, parent2)
            child = successor(child)
            next_generation.append(child)
        population = next_generation
    fitness_scores = [fitness_function(path, distances, regions) for path in population]
    best_index = np.argmax(fitness_scores)
    return population[best_index], fitness_scores[best_index]

# Example ACO paths to initialize the population and test the genetic algorithm
aco_paths = [["R1", "R2", "R3", "R4", "R5"], ["R1", "R3", "R2", "R5", "R4"]]

# Running the Genetic Algorithm with the defined functions
best_path, best_fitness = genetic_algorithm(aco_paths, num_generations=10, num_paths=6, distances=distances, regions=regions)

print("Best Path:", best_path)
print("Best Fitness Score:", best_fitness)
```

##### Initialize Population: Generate an initial set of potential paths for agents.
##### Fitness Evaluation: Evaluate each path based on the total travel time and the number of regions covered.
##### Selection: Choose the fittest paths to carry forward.
##### Crossover: Combine segments of paths to explore new route possibilities.
##### Mutation: Introduce minor changes to paths to avoid local optima.
##### Expected Output: Optimized paths with minimized travel time, ensuring all regions are covered.

Start with total number of units of agents at `path[0]`. For each region `path[i]`, the number units of agents assigned can be from the available units at that point down to one. Compute all possible allocations for the regions (`path[1]`, `path[2]`, `path[3]`, `path[4]`) based on the remaining units after allocation at the current region. Recursively list all possible combinations of allocations along the path.

```python

def utility_function(regions, path):
    """
    Calculates utility based on the remaining unweighted damage in each region.
    """
    total_damage = sum(regions[region]["fires"] + regions[region]["roads_blocked"] for region in path)
    return total_damage

def allocate_units(fire_units, police_units, region_damage, max_units_per_region):
    allocations = []
    for fire_alloc in range(0, min(fire_units, max_units_per_region) + 1):
        for police_alloc in range(0, min(police_units, max_units_per_region - fire_alloc) + 1):
            new_fire_damage = max(0, region_damage["fires"] - fire_alloc * 10)
            new_blockage_damage = max(0, region_damage["roads_blocked"] - police_alloc * 10)
            allocations.append((fire_alloc, police_alloc, fire_units - fire_alloc, police_units - police_alloc, {"fires": new_fire_damage, "roads_blocked": new_blockage_damage}))
    return allocations

def minimax(path, regions, remaining_fire_units, remaining_police_units, depth, alpha, beta, maximizing_player, allocation_path=None, max_units_per_region=5):
    if allocation_path is None:
        allocation_path = []
    if depth == len(path) or (remaining_fire_units <= 0 and remaining_police_units <= 0):
        region_path = path[:depth]  
        return utility_function(regions, region_path), allocation_path[:]

    region = path[depth]
    region_damage = {"fires": regions[region]["fires"], "roads_blocked": regions[region]["roads_blocked"]}
    best_allocation = []
    
    if maximizing_player:
        min_damage = float('inf')
        for fire_alloc, police_alloc, new_remaining_fire, new_remaining_police, new_damage in allocate_units(remaining_fire_units, remaining_police_units, region_damage, max_units_per_region):
            original_damage = {"fires": regions[region]["fires"], "roads_blocked": regions[region]["roads_blocked"]}
            regions[region].update(new_damage)
            allocation_path.append((region, fire_alloc, police_alloc))
            eval_score, best_sub_allocation = minimax(path, regions, new_remaining_fire, new_remaining_police, depth + 1, alpha, beta, False, allocation_path, max_units_per_region)
            
            if eval_score < min_damage:
                min_damage = eval_score
                best_allocation = best_sub_allocation
            
            alpha = min(alpha, eval_score)
            regions[region].update(original_damage)
            allocation_path.pop()
            if beta <= alpha:
                break
        return min_damage, best_allocation
    else:
        max_damage = float('-inf')
        for region in path[:depth+1]:
            regions[region]["fires"] = min(100, regions[region]["fires"] + 10)
            regions[region]["roads_blocked"] = min(100, regions[region]["roads_blocked"] + 10)
        
        eval_score, best_sub_allocation = minimax(path, regions, remaining_fire_units, remaining_police_units, depth, alpha, beta, True, allocation_path, max_units_per_region)
        
        for region in path[:depth+1]:
            regions[region]["fires"] = max(0, regions[region]["fires"] - 10)
            regions[region]["roads_blocked"] = max(0, regions[region]["roads_blocked"] - 10)
        
        max_damage = max(max_damage, eval_score)
        beta = max(beta, eval_score)
    
        return max_damage, best_sub_allocation

# Initialize parameters
path = ["R1", "R2", "R3", "R4", "R5"]
total_fire_units = 8
total_police_units = 6
max_units_per_region = 3

# Run Minimax with Alpha-Beta Pruning and track best path
best_allocation_score, best_allocation_path = minimax(
    path, regions, total_fire_units, total_police_units, depth=0, alpha=float('-inf'), beta=float('inf'), maximizing_player=True, max_units_per_region=max_units_per_region)

print("Best Allocation Score (Remaining Damage):", best_allocation_score)
print("Best Allocation Path (Region, Fire Units Allocated, Police Units Allocated):", best_allocation_path)
```

##### Impact: The Game Theory analysis ensures that the actions of fire trucks and police do not counteract each other.
```python
# Define parameters and actions for each player
fire_truck_actions = list(range(0, 9))  # Fire trucks can allocate 0 to 8 units
police_actions = list(range(0, 7))  # Police can allocate 0 to 6 units

# Initialize the payoff matrix
payoff_matrix = [[0 for _ in range(len(police_actions))] for _ in range(len(fire_truck_actions))]

# Function to simulate the rescue operation and calculate remaining damage for each action pair
def calculate_payoff(fire_units, police_units, regions):
    """
    Simulates the outcome for a given allocation of fire and police units and returns the remaining total damage.
    """
    # Copy the regions dictionary to avoid modifying the original data
    simulated_regions = {region: data.copy() for region, data in regions.items()}
    
    # Apply fire trucks' effect on fire damage
    for region, data in simulated_regions.items():
        data["fires"] = max(0, data["fires"] - fire_units * 10)  # Each fire truck reduces fire by 10%
    
    # Apply police effect on road blockages
    for region, data in simulated_regions.items():
        data["roads_blocked"] = max(0, data["roads_blocked"] - police_units * 10)  # Each police unit reduces blockage by 10%
    
    # Calculate remaining damage
    total_remaining_damage = sum(data["fires"] + data["roads_blocked"] for data in simulated_regions.values())
    return total_remaining_damage

# Populate the payoff matrix by evaluating each combination of actions
for i, fire_units in enumerate(fire_truck_actions):
    for j, police_units in enumerate(police_actions):
        payoff_matrix[i][j] = calculate_payoff(fire_units, police_units, regions)

# Print the payoff matrix
print("Payoff Matrix (Remaining Damage):")
for row in payoff_matrix:
    print(row)

# Identify optimal strategies (Nash Equilibrium Analysis)
min_damage = float('inf')
optimal_fire_units = 0
optimal_police_units = 0

for i in range(len(fire_truck_actions)):
    for j in range(len(police_actions)):
        if payoff_matrix[i][j] < min_damage:
            min_damage = payoff_matrix[i][j]
            optimal_fire_units = fire_truck_actions[i]
            optimal_police_units = police_actions[j]


def calculate_distributed_payoff(fire_units_distribution, police_units_distribution, regions):
    """
    Simulates the outcome for a given distribution of fire and police units across regions.
    Returns the remaining total damage.
    """
    # Create a copy of regions to avoid modifying the original data
    simulated_regions = {region: data.copy() for region, data in regions.items()}
    
    # Apply fire trucks' effect on fire damage for each region according to distribution
    for i, (region, data) in enumerate(simulated_regions.items()):
        fire_units = fire_units_distribution[i]
        data["fires"] = max(0, data["fires"] - fire_units * 10)  # Each fire truck reduces fire by 10%
    
    # Apply police effect on road blockages for each region according to distribution
    for i, (region, data) in enumerate(simulated_regions.items()):
        police_units = police_units_distribution[i]
        data["roads_blocked"] = max(0, data["roads_blocked"] - police_units * 10)  # Each police unit reduces blockage by 10%
    
    # Calculate remaining damage
    total_remaining_damage = sum(data["fires"] + data["roads_blocked"] for data in simulated_regions.values())
    return total_remaining_damage

            
def calculate_nash_equilibrium(fire_truck_actions_per_region, police_actions_per_region, regions):
    """
    Function to evaluate Nash Equilibrium by iterating over possible actions of fire trucks and police units.
    Determines the best equilibrium point that minimizes remaining damage across all regions.
    """
    # Initialize a payoff matrix based on possible distributed unit allocations
    payoff_matrix = [[0 for _ in range(len(police_actions_per_region))] for _ in range(len(fire_truck_actions_per_region))]

    # Calculate distributed payoff for each combination of fire truck and police unit actions
    for i, fire_units in enumerate(fire_truck_actions_per_region):
        for j, police_units in enumerate(police_actions_per_region):
            fire_distribution = [fire_units] * 5  # Assume equal distribution for simplicity
            police_distribution = [police_units] * 5  # Assume equal distribution for simplicity
            payoff_matrix[i][j] = calculate_distributed_payoff(fire_distribution, police_distribution, regions)
    
    # Find Nash Equilibrium by identifying the minimal row and column values
    nash_fire_units, nash_police_units = None, None
    min_damage = float('inf')
    
    for i, fire_units in enumerate(fire_truck_actions_per_region):
        for j, police_units in enumerate(police_actions_per_region):
            current_damage = payoff_matrix[i][j]
            if all(payoff_matrix[fi][j] >= current_damage for fi in range(len(fire_truck_actions_per_region))) and \
               all(payoff_matrix[i][pj] >= current_damage for pj in range(len(police_actions_per_region))):
                if current_damage < min_damage:
                    min_damage = current_damage
                    nash_fire_units = fire_units
                    nash_police_units = police_units

    if nash_fire_units is not None and nash_police_units is not None:
        return nash_fire_units, nash_police_units, min_damage
    else:
        return None, None, None  # No Nash Equilibrium found

    
print(f"Optimal Fire Truck Units: {optimal_fire_units}")
print(f"Optimal Police Units: {optimal_police_units}")
print(f"Minimum Remaining Damage: {min_damage}")
```

## Bayesian Networks for Uncertain Inferences

#### Bayesian Network for Inference on Rescue Success Probability
##### Purpose: The Bayesian Network models the probabilistic dependencies between conditions such as aftershocks, road blockages, and fire spread to predict the likelihood of a successful rescue operation. This is particularly useful when dealing with uncertainties in the scenario.
##### Impact: By calculating the probability of successful rescues given current conditions, the Bayesian Network allows responders to prioritize regions with the highest chance of successful mitigation. It also helps strategize in cases where aftershocks and road blockages heavily influence rescue effectiveness.

```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
import networkx as nx
import matplotlib.pyplot as plt

# Step 1: Define the Bayesian Network Structure
bn_model = BayesianNetwork([
    ("Aftershocks", "Road Blockages"),
    ("Aftershocks", "Fire Spread"),
    ("Road Blockages", "Travel Time"),
    ("Fire Spread", "Rescue Success"),
    ("Road Blockages", "Rescue Success"),
])

# Plot the Bayesian Network structure
plt.figure(figsize=(8, 6))
G = nx.DiGraph()
G.add_edges_from([
    ("Aftershocks", "Road Blockages"),
    ("Aftershocks", "Fire Spread"),
    ("Road Blockages", "Travel Time"),
    ("Fire Spread", "Rescue Success"),
    ("Road Blockages", "Rescue Success"),
])
nx.draw_networkx(G, with_labels=True, node_color='skyblue', font_weight='bold', edge_color='gray', node_size=3000)
plt.title("Bayesian Network Structure for Rescue Scenario")
plt.show()

# Step 2: Define Conditional Probability Tables (CPTs) for each node

# Aftershocks: Prior probability of aftershocks occurring
cpd_aftershocks = TabularCPD(
    variable="Aftershocks", variable_card=2,
    values=[[0.8],  # No aftershocks
            [0.2]]  # Aftershocks occur
)

# Road Blockages: Conditional on Aftershocks
cpd_road_blockages = TabularCPD(
    variable="Road Blockages", variable_card=2,
    values=[[0.9, 0.4],  # No road blockages given no aftershocks / aftershocks
            [0.1, 0.6]],  # Road blockages given no aftershocks / aftershocks
    evidence=["Aftershocks"], evidence_card=[2]
)

# Fire Spread: Conditional on Aftershocks
cpd_fire_spread = TabularCPD(
    variable="Fire Spread", variable_card=2,
    values=[[0.85, 0.3],  # No fire spread given no aftershocks / aftershocks
            [0.15, 0.7]],  # Fire spread given no aftershocks / aftershocks
    evidence=["Aftershocks"], evidence_card=[2]
)

# Travel Time: Conditional on Road Blockages
cpd_travel_time = TabularCPD(
    variable="Travel Time", variable_card=2,
    values=[[0.7, 0.3],  # Short travel time if no blockages / blockages
            [0.3, 0.7]],  # Long travel time if no blockages / blockages
    evidence=["Road Blockages"], evidence_card=[2]
)

# Rescue Success: Conditional on Fire Spread and Road Blockages
cpd_rescue_success = TabularCPD(
    variable="Rescue Success", variable_card=2,
    values=[[0.95, 0.6, 0.6, 0.2],  # Successful rescue given all combinations of fire spread / blockages
            [0.05, 0.4, 0.4, 0.8]],  # Failed rescue given all combinations of fire spread / blockages
    evidence=["Fire Spread", "Road Blockages"], evidence_card=[2, 2]
)

# Step 3: Add the CPDs to the model
bn_model.add_cpds(cpd_aftershocks, cpd_road_blockages, cpd_fire_spread, cpd_travel_time, cpd_rescue_success)
print("Bayesian Network is valid:", bn_model.check_model())

# Step 4: Perform inference on the Bayesian Network
inference = VariableElimination(bn_model)

# Query the probability of a successful rescue given specific conditions
query_result = inference.query(variables=["Rescue Success"], evidence={"Aftershocks": 1, "Road Blockages": 1})
print("Probability of Successful Rescue given Aftershocks and Road Blockages:", query_result)

# Further optimization queries
query_result_optimal = inference.query(variables=["Rescue Success"], evidence={"Fire Spread": 0, "Road Blockages": 0})
print("Probability of Successful Rescue with No Fire Spread and No Road Blockages:", query_result_optimal)
```
